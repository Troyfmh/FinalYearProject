```python
## Assign amino acids sdf to variable amino   
amino = Chem.SDMolSupplier("AminoAcids.sdf")
```

```python
## See the length of the amino sdf file
len(amino)
```

```python
## For every molecule (mol) in the sdf amino, get the number of atoms and print it
for mol in amino:
    x = mol.GetNumAtoms()
    print(x)
```

```python
## So by breaking the for loop when x reaches a certain number we can stop it from running all 1517 images and only generate a few
## allowing us to see whats in the amino acid data set. 

x = 10
for mol in amino:        
    img = Draw.MolToImage(mol)
    ## Display img is much better than img.show as it runs in the vsc console and not on the photos file on the computer 
    display(img)
    x = x + 1 
    if x >= 11:
        break

```

```python
## Now I want to get the smile strings so i can search easily for if an amino acid contains a suflur molecule

#for mol in amino:
#    smile = Chem.MolToSmiles(mol)
#    print(smile)


#This code outputs the smile string only if it has an S in it 
for mol in amino:
    smile = Chem.MolToSmiles(mol)
    if 'S' in smile:
        print(smile)
        #And we can also draw them using the following command
        img = Draw.MolToImage(mol)
        display(img)
```
